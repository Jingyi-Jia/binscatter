```python
import numpy as np
import pandas as pd
# Generate a fake dataset
np.random.seed(333)
n = 1000
x = np.random.uniform(-10, 10, size=n)
y = 3 * x + np.random.normal(0, 5, size=n)
z = np.random.uniform(-5, 5, size=n)
ind = np.random.choice(['A', 'B', 'C', 'D'], size=n)

# Create a DataFrame
df = pd.DataFrame({'y': y, 'x': x, 'z':z,'ind': ind})

# Save the DataFrame as a CSV file
df.to_csv('extensive_fake_data.csv', index=False)
df.head()
```

```python
# Load the dataset
df = pd.read_csv('extensive_fake_data.csv')

from binscatter.binscatter import create_binscatter
# Binscatter plot
create_binscatter(df['x'], df['y']);
# Binscatter plot with fixed effects
create_binscatter(df['x'], df['y'], fixed_effects=df['ind']);
create_binscatter(df['x'], df['y'], controls= df['z']);
```
